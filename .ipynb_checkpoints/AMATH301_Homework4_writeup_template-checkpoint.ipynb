```python
import time

p = [2, 3]
tol = 1e-9 # Set the tolerance - note that it changes
# Now run gradient descent!

gradf_xy = lambda x,y: np.array([4*x**3 - 42*x + 4*x*y + 2*y**2 - 14,
                                 4*y**3 - 26*y + 4*x*y + 2*x**2 - 22])
gradf = lambda p: gradf_xy(p[0], p[1])

def gradDes1(p):
    for k in range(8000):
        grad = gradf(p)
        if np.linalg.norm(grad)<tol:
            break
        
        phi = lambda t: p - t*grad
        f_of_phi = lambda t: f(phi(t)) # Create a function of "heights along path"
        tmin = scipy.optimize.fminbound(f_of_phi,0,1) # Find time it takes to reach min height
        p = phi(tmin); # Find the point on the path and update your guess
    return [p, k]


t1 = time.time()
gD1p = gradDes1(p)
t2 = time.time()
print(gD1p)
print(t2-t1)
```

### Part f - discussion

- I found that the Gradient descent algorithm....
Did not always converge. It didn't converge when the tstep got too big, in this case only when it was .025. This is because the gradient descent probably kept overshooting back and forth between either side of the true value.

- I found that Gradient Descent converged fastest with...
The tstep values of .01 and .02. These were faster than tstep .025 because that one never made it. The tstep .01 and .02 both converged in relatively few iterations, and the iterations were quick. fminbound, however, has to do too much calulation for each iteration so it takes longer.

- I found that Gradient Descent converged with the fewest iterations with...
fminbound converged in the fewest iterations because with each iteration it covers the most distance that is useful. However, this takes longer because of the calculations to do this. The tstep works faster, but more iterations that are fast.

- The tsteps are generally faster if you choose an appropriate step size, but they take more iterations.
